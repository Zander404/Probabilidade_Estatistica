```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("dados_estudantes.csv", sep=";")
```

```python
df.head(5)
```

```python
df3 = df2 = pd.DataFrame()
df2["Altura"] = df["Alt"]
df2["Peso"] = df["Peso"]

df3["Idade"] = df["Idade"]
df3["Altura"] = df["Alt"]
```

```python
plt.style.use('_mpl-gallery')
fig, ax = plt.subplots(1, 2, figsize=(12,6))

# Comparo entre ALTURA X PESO
ax[0] = sns.scatterplot(df2, x="Altura", y="Peso", hue="Altura", legend="auto", ax=ax[0])

# Comparo entre ALTURA X IDADE
ax[1] = sns.scatterplot(df3, x="Altura", y="Idade", hue="Altura", legend="auto", ax=ax[1])
plt.show()
```
